```python
#age
```

```python
Q1=np.percentile(data2['Age'],25,method='midpoint')
Q2=np.percentile(data2['Age'],50,method='midpoint')
Q3=np.percentile(data2['Age'],75,method='midpoint')
print("Q1 IS:",Q1)
print("Q2 IS:",Q2)
print("Q3 IS:",Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
low_limit=Q1-1.5*IQR
Up_limit= Q3+1.5*IQR
print("lowe value: ",low_limit)
print("upper value: ",Up_limit)
```

```python
outliers=[]

for x in data2['Age']:
  if((x<low_limit)or (x>Up_limit)):
    outliers.append(x)


  print(outliers)
```

```python
#removing outliers
```

```python
data2['Age']=data2['Age'].clip(lower=low_limit,upper=Up_limit)
```

```python
outliers=[]

for x in data2['Age']:
  if((x<low_limit)or (x>Up_limit)):
    outliers.append(x)


  print(outliers)
```
